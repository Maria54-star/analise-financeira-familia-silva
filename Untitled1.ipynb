```python
# Análise de Gastos Domésticos - Água, Luz e Internet
# Projeto criado para portfólio no GitHub

import matplotlib.pyplot as plt

# ---- DADOS ----
categorias = ['Água', 'Luz', 'Internet']
valores = [80, 150, 120]  # valores em R$ - você pode alterar!

# ---- GRÁFICO DE PIZZA ----
cores = ['#1E90FF', '#FFD700', '#32CD32']
explode = (0.05, 0.05, 0.05)  # destaca as fatias

plt.figure(figsize=(7, 7))
plt.pie(
    valores,
    labels=categorias,
    colors=cores,
    explode=explode,
    autopct='%1.1f%%',
    startangle=140,
    shadow=True
)

plt.title('Distribuição de Gastos Domésticos\nÁgua | Luz | Internet', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('gastos_domesticos.png', dpi=150)
plt.show()

print("\n💰 Resumo dos Gastos:")
for cat, val in zip(categorias, valores):
    print(f"   {cat}: R$ {val:.2f}")
print(f"   TOTAL: R$ {sum(valores):.2f}")
```

```python
# =====================================================
# Análise Financeira Completa - Família Silva
# Família carioca de 4 pessoas | Janeiro/2026
# Projeto de portfólio | Autor: (Maria Claudia)
# =====================================================

import matplotlib.pyplot as plt

# ---- SALÁRIO DA FAMÍLIA ----
salario_pai = 2500.00
salario_mae = 2000.00
salario_total = salario_pai + salario_mae

# ---- GASTOS FIXOS (água, luz, internet) ----
gastos_fixos = {
    'Água': 95.00,
    'Luz': 210.00,
    'Internet': 149.90
}

# ---- GASTOS EXTRAS ----
gastos_extras = {
    'Alimentação': 800.00,
    'Transporte': 350.00,
    'Escola dos Filhos': 400.00,
    'Lazer / Passeios': 200.00,
    'Saúde / Plano de Saúde': 320.00
}

# ---- CÁLCULOS ----
total_fixos = sum(gastos_fixos.values())
total_extras = sum(gastos_extras.values())
total_gastos = total_fixos + total_extras
saldo = salario_total - total_gastos

# ---- RESUMO NO CONSOLE ----
print("=" * 50)
print("   ANÁLISE FINANCEIRA - FAMÍLIA SILVA")
print("   Rio de Janeiro | Janeiro/2026")
print("   👨‍👩‍👧‍👦 Família de 4 pessoas")
print("=" * 50)
print(f"\n💼 SALÁRIO COMBINADO: R$ {salario_total:,.2f}")
print(f"   Pai: R$ {salario_pai:,.2f}")
print(f"   Mãe: R$ {salario_mae:,.2f}")

print(f"\n🏠 GASTOS FIXOS: R$ {total_fixos:,.2f}")
for item, val in gastos_fixos.items():
    print(f"   {item}: R$ {val:,.2f}")

print(f"\n📋 GASTOS EXTRAS: R$ {total_extras:,.2f}")
for item, val in gastos_extras.items():
    print(f"   {item}: R$ {val:,.2f}")

print(f"\n💰 TOTAL DE GASTOS: R$ {total_gastos:,.2f}")
if saldo >= 0:
    print(f"✅ SALDO FINAL: R$ {saldo:,.2f}")
else:
    print(f"⚠️  SALDO NEGATIVO: R$ {saldo:,.2f}")
print("=" * 50)

# ---- GRÁFICO 1: PIZZA - Todos os gastos ----
todos_gastos = {**gastos_fixos, **gastos_extras}
labels = list(todos_gastos.keys())
valores = list(todos_gastos.values())

cores = ['#1E90FF', '#FF6B35', '#32CD32',
         '#FF4500', '#FFD700', '#9370DB',
         '#20B2AA', '#FF69B4']

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# Pizza
axes[0].pie(
    valores,
    labels=labels,
    colors=cores[:len(labels)],
    autopct='%1.1f%%',
    startangle=140,
    shadow=True,
    textprops={'fontsize': 10}
)
axes[0].set_title('Distribuição de Todos os Gastos\nFamília Silva - Jan/2026',
                   fontsize=13, fontweight='bold')

# ---- GRÁFICO 2: BARRAS - Salário x Gastos ----
categorias_bar = ['Salário\nCombinado', 'Gastos\nFixos', 'Gastos\nExtras', 'Total\nGastos', 'Saldo\nFinal']
valores_bar = [salario_total, total_fixos, total_extras, total_gastos, saldo]
cores_bar = ['#2ECC71', '#3498DB', '#E67E22', '#E74C3C', '#27AE60' if saldo >= 0 else '#C0392B']

bars = axes[1].bar(categorias_bar, valores_bar, color=cores_bar, edgecolor='white', linewidth=1.2)
axes[1].set_title('Salário x Gastos - Família Silva\nJaneiro/2026',
                   fontsize=13, fontweight='bold')
axes[1].set_ylabel('Valor em R$', fontsize=11)
axes[1].set_ylim(0, max(valores_bar) * 1.2)

for bar, val in zip(bars, valores_bar):
    axes[1].text(bar.get_x() + bar.get_width() / 2,
                 bar.get_height() + 30,
                 f'R$ {val:,.0f}',
                 ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.suptitle('Análise Financeira - Família Silva | Rio de Janeiro',
             fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('analise_financeira_familia_silva.png', dpi=150, bbox_inches='tight')
plt.show()
```
